```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#Call out Player
player = input("Enter Player's Name: ")

#read in data
df = pd.read_excel('/Users/mikeysoukup/Desktop/Python/Accuracy/FT%.xlsx','{}'.format(player))

#Create regression plot
season = list(df.index)
ftp = list(df['FT%'])
slope, intercept, r_value, p_value, std_err = stats.linregress(season,ftp)
sns.set_style("darkgrid")
plt.figure(figsize = (14,7))
sns.regplot(x = season, y = ftp,
            scatter_kws={"s": 75},
           line_kws={'label': "$R^2:{0:.4f}$".format(round(float(r_value**2),4))})
plt.title('{} Career FT% Over Season in League'.format(player),fontsize = 'x-large')
plt.xlabel('Year in League',fontsize = 'large')
plt.ylabel('FT%',fontsize = 'large')
plt.legend(prop={'size': 15})
plt.xticks(season,fontsize = 'large')
plt.yticks(fontsize = 'large')
#plt.savefig('{} FTp Over Career'.format(player))
plt.show()
```
