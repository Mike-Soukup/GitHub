```python
import matplotlib.pyplot
from matplotlib.pyplot import *
import numpy
from numpy import linspace
#Get necessary software packages into the Jupyter Notebook
```

3x + 10y = 115

11x + 4y = 95

```python
x = linspace(-1,15,100) #linspace will create 100 points between -1 and 15, this is our x range for the graph
y1 = (115-3*x)/10
y2 = (95-11*x)/4
```

```python
xlabel('x-axis')
ylabel('y-axis')
plot(x,y1,'r')
plot(x,y2,'b')
legend(('3x+10y=115','11x+4y=95'),loc = 3)
title('Solving a System of Equations with a Unique Solution')
show()
```

```python
a = linspace(0,6,100)
b1 = (6-2*a)/-3
b2= (8+4*a)/6
figure()
plot(a,b1)
plot(a,b2)
xlabel('a-axis')
ylabel('b-axis')
title('Showing an Inconsistent System of Equations')
show()
```

```python
x = linspace(0,2,100)
y1 = 2.0-x**2
y2 = x**2
figure()
plot(x,y1)
plot(x,y2)
xlabel('x-axis')
ylabel('y-axis')
title('Consistent system of non-linear equations')
show()
```

```python
x = linspace(0,100,100)
y1 = x*x+x
y2 = y1+1000
figure()
plot(x,y1,c='b')
plot(x,y2,c='r')
xlabel('x-axis')
ylabel('y-axis')
legend(('y1=x*x+x','y2=x*x+x+1000'),loc='best')
title('Non-linear system of Inconsistent equations')
show()
```
