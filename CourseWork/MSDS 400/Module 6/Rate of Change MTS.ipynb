```python
import numpy
from numpy import arange, cos
import matplotlib.pyplot
from matplotlib.pyplot import *
```

```python
def der(x, delta):
    delta = float(delta)
    if delta < 0.0000001:
        print('Value chose for delta is too small.')
        return 1 / delta
    else:
        slope = (f(x+delta)-f(x))/delta
        return slope
    
```

```python
def f(x):
    f = cos(x)
    return f

point = 1.0
number = 510
increment = 10
y = []
x = []
```

```python
for k in range(increment, number, increment):
    delta = 1.0/(k+1)
    d = der(point,delta)
    print("{:10}{:10}".format(k,round(d,3)))
    x = x + [k]
    y = y + [d]
    max_x = k + increment

limit = der(point, 0.0000001)
print('Final value equals')
limit
```

```python
figure()
xlim(0,max_x+50)
ylim(min(y)-0.05,max(y)+0.05)

scatter(540,limit,color='g',s=40,label = 'limiting slope')
legend(['limiting slope'],loc = 'best')
scatter(x,y,c='k',s=20)
title('Example of Convergence to Instantaneous Rate of Change')
xlabel('x-axis')
ylabel('y-axis')
plot(x[0:50],y[0:50])
show()
```

```python
w = arange(point-1.0,point+1.1,0.1)
t = f(point)+limit*(w-point)

domain = 3.14
u = arange(point-domain,point+domain+0.1,0.1)
z = f(u)
```

```python
figure()
xlim(point-domain-.1,point+domain+0.1)
ylim(max(z)+0.5,min(z)-0.5)

plot(w,t,c='r')
plot(u,z,c='b')
scatter(point,f(point),c='g',s=40)
xlabel('x-axis')
ylabel('y-axis')
title('Plot showing function and tangent at a point')
show()
```

```python
#Lial Section 11.3 Examples 3-5
#create a function to calculate the derivative and find the derivative at the points listed in the examples and plot

def derivative(x,function):
    h = 0.000000001
    m = (function(x+h)-function(x))/h
    return m

def ff(x):
    ff = 2*(x**2)-5*x+40
    return ff

a_slope = derivative(4,ff)

figure()

t = arange(2,8,0.05)
s = ff(t)
y = (a_slope*t)+8

xlabel('x-axis')
ylabel('y-axis')
plot(t,s,color = 'b')
plot(t,y,color = 'k')
scatter(4,ff(4),color='r',s = 20)
title('Example 3: Velocity')
legend(['Position','Slope is ~{}'.format(round(a_slope,2))])
show()
```

```python
#Example #4

def C(x):
    C = 100 + 15*x - x**2
    return C

slope_one_case = derivative(1,C)

x = arange(-2,4,0.05)
C_x = C(x)
line = (derivative(1,C)*x)+101
figure()
plot(x,C_x,color='b')
plot(x,line,color='k')
scatter(1,C(1),color='r')
legend(['C(x)','Slope ~{}'.format(round(derivative(1,C),2))])
title('Example 4 Manufacturing')
xlabel('x-axis')
ylabel('y-axis')
show()
```
