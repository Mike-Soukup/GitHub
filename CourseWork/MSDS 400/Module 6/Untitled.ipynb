```python
import matplotlib.pyplot
from matplotlib.pyplot import *
import numpy 
from numpy import *

x = arange(-10,10,.1)
f_x = (3*x+2)/(2*x+4)

ymin = -5
ymax = 5
xmin = -5
xmax = 5

figure()
vlines(0,ymin,ymax)
hlines(0,xmin,xmax)

plot(x,f_x)

show()

#approach from left

```

```python
x_left = arange(-3,-2,.001)
x_left = list(x_left)

for i in x_left:
    f_x = (3*i+2)/(2*i+4)
    print("{:10.3}{:10.3}".format(i,f_x))
```

```python
x_right = arange(-1,-2,-.001)
x_right = list(x_right)

for i in x_right:
    f_x = (3*i+2)/(2*i+4)
    print("{:10.3}{:10.3}".format(i,f_x))
```
