# Trapezoidal Rule for Numerical Integration

```python

def integrate_trap(a, b, n):
    total = 0.0
    delta = (b-a)/n
    i = 0
    while i < n:
        total = total + delta * ((f(a + delta * (i + 1)) + f(a + delta * i)) / 2.0) #Trapezoidal rule
        i = i + 1
    return total


```

# Midpoint Rule for Numerical Integration

```python
def integrate_midpoint(a,b,n):
    total = 0.0
    delta = (b-a)/n
    i = 0
    while i < n:
        total = total + (delta * f(a + delta*(0.5 + i)))
        i = i + 1
    return total
```

```python
def k(x):
    return (2*math.pi)*(1/x)*math.sqrt(1+(1/x**4))

def integrate_midpoint(a,b,n):
    total = 0.0
    delta = (b-a)/n
    i = 0
    while i < n:
        total = total + (delta * k(a + delta*(0.5 + i)))
        i = i + 1
    return total

b = 1
a = [10,100,1000,10000,100000,1000000,10000000,100000000]
n = 10000
t = []

for i in a:
    t = t + [integrate_midpoint(b,i,n)]
t
```

# Simpson's Rule for Integration AND X (rand var) to Z-score function

```python
# This is the density function for the standard normal distribution
def f(inp):
    return (math.exp(-inp * inp / 2)) / math.sqrt(2.0 * math.pi)

# Simpson's rule is used to integrate over [a,b].

def integrate_simp(a,b,delta):
    total = 0.0
    i = 0
    n = int(float((b-a) / delta))
    if b == a:
        return
    else:
        while i < n:
            x0 = a + delta*i
            x1 = x0 + delta/2
            x2 = x0 + delta
            total = total + (delta*(f(x0) + 4.0*f(x1) + f(x2)) / 6) #why 6 and not 3?
            i = i + 1
        return total

def x_to_z(x,sig,mu):
    z = (x-mu)/sig
    print('z is {}'.format(z))
    return x
x_to_z(70,16,75)
```

# Graph of CDF and Normal distribution for entered value of std. dev for normal curve

```python
nsamples = 100
sample = random.random(nsamples) # this draws a random sample, nsamples form 0-1

nbins = 10 # this defines the number of subintervals for the histogram
bns = float(nbins)

expected = 1.0 / bns #This defines the expected subinterval proportion.

ind = arange(nbins) #This sets ind to serve as a range of indices.
h = [0] * nbins #This prepares h ot serve as a list of the proper length
histogram = {} #This defines histogram as a void dictionary.

for k in ind:
    histogram[k] = 0 #This initializes the dictionary with zero counts.

for v in sample: #Run v across each random sample and place it into its correct bucket in histogram
    for k in ind:
        xk = float(k)
        if xk / bns <= v < (xk+1)/bns:
            histogram[k] += 1

for k in ind:  #Turn counts into proportions
    x = histogram[k]
    x = x / float(nsamples)
    h[k] = [x]
    histogram[k] = x

#Measure degree of convergence of histogram to limiting uniform distribution with sum of absolute differences between
#the expected proportion and the observed proportion

total = 0.0
for k in ind:
    total = total + abs(expected - histogram[k])
total = format(total, '0.4e')
print('Sum of Absolute Differences = {}'.format(total))

h = swapaxes(h,0,1)[0] # Swapping x/y axes for pyplot
cell = ind + 0.5 #This will center the bar in the middle of the subinterval
plt.figure()
plt.bar(cell, h, width=0.5, align = 'center', color = 'r')
plt.plot(cell,h)

string = str(nsamples) + ' Absolute Difference = ' + str(total)
plt.title('Histogram n =' + string)

plt.ylabel('Proportions')
plt.xlabel('Subintervals')
plt.show()
```
