```python
af30d = af30d.drop(columns = ['Shrinkage','Pinholes','Comments','Color Grade','0-10','11-20','131-140','Yards Made','Yards Produced','Sum'])
af30d.head()
```

```python
af30d['Run Order'] = af30d['Run Order'].astype(str)
af30d.info()
```

```python
#Set up least squares functions
def least_squares_m(x,y):
    n = len(x)
    xy = []
    for i in range(len(x)):
        xy = xy + [x[i]*y[i]]
    x_sq = []
    for i in range(len(x)):
        x_sq = x_sq + [x[i]**2]
    sum_xy = sum(xy)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x_sq = sum(x_sq)
    m = ((n*sum_xy)-(sum_x*sum_y))/((n*sum_x_sq)-(sum_x**2))
    return m
def least_squares_b(x,y,m):
    sum_y = sum(y)
    sum_x = sum(x)
    n = len(x)
    b = (sum_y - (m*sum_x))/n
    return b
```

```python
af30_rub_lot_col = af30d.loc[:,'Rub Lot']
af30_rub_lot_dic = {}
for i in af30_rub_lot_col:
    af30_rub_lot_dic[i] = 1
af30_rubberlots = af30_rub_lot_dic.keys()
print(af30_rubberlots)
#Set up indexer
k = 0
for j in af30_rubberlots:
    plt.figure(k)
    is_lot = af30d['Rub Lot'] == j
    af30d_lot = af30d[is_lot]
    run_order = af30d_lot['Run Order']
    rub_yield = af30d_lot['Yield [=] %']
    plt.scatter(run_order,rub_yield)
    plt.xlabel('Run Order')
    plt.xticks(rotation = 'vertical')
    plt.ylabel('Yield [=] %')
    plt.title('Rubber RM Lot {} over Time as Run Order'.format(j))
    plt.show()
    k += 1
    
```
