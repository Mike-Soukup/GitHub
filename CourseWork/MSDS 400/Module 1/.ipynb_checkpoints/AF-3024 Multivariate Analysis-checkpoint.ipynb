# AF-3024 Multi-variate Regression Analysis

#### The goal of this analysis is to understand the key influencers on the AF-3024 Unit Cost produced in Springfield. Because this product is being discussed as providing a manufacturing blueprint for the New Core Splice NPI material, a better understanding of the cost implications is needed.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ucdf = pd.read_excel('AF-3024 Unit Cost Multivariate Regression.xlsx','UnitCostData')
ucdf.head()
```

Next, I want to look at the columns I have and eliminate any that will not be useful in the multivariate analysis. 

```python
ucdf.dtypes
```

From the looks of it the only columns we will need are:
- Time Period
- AMN ID
- Overhead/Machine Price
- Personnel Price
- Raw Material Price
- QTY
- Unit Cost

So we will extract these columns from the dataframe into a new dataframe:

```python
ucdf1 = ucdf[['Time Period','AMN ID','Overhead/Machine Price','Personnel Price','Raw Material Price',
              'QTY','Unit Cost']]
ucdf1.head()
```

Next, I want to subset 3 seperate dataframes. One for each of the AMNs:
- 4137007398 or PM-7398
- 4137007399 or PM-7399
- 4191000580 or EC-3025 Jumbo

```python
ucPM7398 = ucdf1['AMN ID']==4137007398
ucPM7398 = ucdf1[ucPM7398]
ucPM7399 = ucdf1['AMN ID']==4137007399
ucPM7399 = ucdf1[ucPM7399]
ucEC3025 = ucdf1['AMN ID']==4191000580
ucEC3025 = ucdf1[ucEC3025]
```

Now I want to take a look at the unit cost distribution for the 3 different SKUs under consideration here:

```python
plt.figure(1)
sns.distplot(ucPM7398['Unit Cost'])
plt.title('Histogram of PM-7398 Unit Cost Data since 2017')
plt.show(1)
plt.figure(2)
sns.distplot(ucPM7399['Unit Cost'])
plt.title('Histogram of PM-7399 Unit Cost Data since 2017')
plt.show(2)
plt.figure(3)
sns.distplot(ucEC3025['Unit Cost'])
plt.title('Histogram of EC-3025 Jumbo Unit Cost Data since 2017')
plt.show(3)
```

All of the Unit Cost Distributions here are Bimodal. This is because there are months/time periods for which these products did not run so the Unit Cost of the product for that month is zero. So I want to filter out these months because it will throw off the analysis.

```python
ucPM7398f = ucPM7398['Unit Cost'] > 0
ucPM7398f = ucPM7398[ucPM7398f]
ucPM7399f = ucPM7399['Unit Cost'] > 0
ucPM7399f = ucPM7399[ucPM7399f]
ucEC3025f = ucEC3025['Unit Cost'] > 0
ucEC3025f = ucEC3025[ucEC3025f]

plt.figure(1)
sns.distplot(ucPM7398f['Unit Cost'])
plt.title('Histogram of PM-7398 Unit Cost Data since 2017')
plt.show(1)
plt.figure(2)
sns.distplot(ucPM7399f['Unit Cost'])
plt.title('Histogram of PM-7399 Unit Cost Data since 2017')
plt.show(2)
plt.figure(3)
sns.distplot(ucEC3025f['Unit Cost'])
plt.title('Histogram of EC-3025 Jumbo Unit Cost Data since 2017')
plt.show(3)
```

This data looks far more normal. Now with these subsetted dataframes, I will begin my multivariate regression analysis.

```python
ucPM7398f = ucPM7398f.drop(columns = ['AMN ID', 'Time Period'])
ucPM7399f = ucPM7399f.drop(columns = ['AMN ID', 'Time Period'])
ucEC3025f = ucEC3025f.drop(columns = ['AMN ID', 'Time Period'])
```

## Pairplot PM-7398

```python
sns.pairplot(ucPM7398f,
            x_vars = ['Raw Material Price','Personnel Price','Overhead/Machine Price','QTY'],
            y_vars = ['Unit Cost'],
            kind = "reg",
            height = 4)
plt.show()

corrpm7398 = ucPM7398f.corr()
rPM7398 = corrpm7398.loc[['Overhead/Machine Price','Personnel Price','Raw Material Price','QTY'],
                         'Unit Cost']
print('The Correlation Coefficients for respective variables to PM-7398 Unit Cost are: \n \n', rPM7398)
```

PM-7398 shows a strong positive correlation between Unit Cost and Overhead/Machine Price. This seems to be the dominant factor on the PM-7398 unit cost. 

There are slightly positive correlations between Unit Cost and Personnel Price along with Raw Material Price, but not as strong as Overhead/Machine Price

There is a weak negative correlation between Unit Cost and QTY.

## Pairplot PM-7399

```python
sns.pairplot(ucPM7399f,
            x_vars = ['Raw Material Price','Personnel Price','Overhead/Machine Price','QTY'],
            y_vars = ['Unit Cost'],
            kind = "reg",
            height = 4)
plt.show()

corrpm7399 = ucPM7399f.corr()
rPM7399 = corrpm7399.loc[['Overhead/Machine Price','Personnel Price','Raw Material Price','QTY'],
                         'Unit Cost']
print('The Correlation Coefficients for respective variables to PM-7399 Unit Cost are: \n \n', rPM7399)
```

PM-7399 shows similar correlations between Unit Cost and the four variables choosen as PM-7398. There is a strong positive correlation with Overhead/Machine Price followed by a moderate positive correlation with Personnel Price and then weak positive correlation with Raw Material Price.

There is a weak negative correlation between Unit Cost and QTY.

## Pairplot EC-3025 Jumbo

```python
sns.pairplot(ucEC3025f,
            x_vars = ['Raw Material Price','Personnel Price','Overhead/Machine Price','QTY'],
            y_vars = ['Unit Cost'],
            kind = "reg",
            height = 4)
plt.show()

correc3025 = ucEC3025f.corr()
rEC3025 = correc3025.loc[['Overhead/Machine Price','Personnel Price','Raw Material Price','QTY'],
                         'Unit Cost']
print('The Correlation Coefficients for respective variables to EC-3025 Jumbo Unit Cost are: \n \n', rEC3025)
```

The EC-3025 Jumbo data shows much stronger correlations with the 4 variables analyzed. The Personnel Price and Overhead/Machine Price seem to have similar strength positive correlations with Unit Cost. The Raw Material Price has a more pronounced positive correlation compared to the PMs and there is a more defined negative correlation with QTY.

This indicates that the run length is having an impact on EC-3025 jumbo Unit Cost compared to the PMs made on the moguls and pony mixers.

```python
ucbdf = pd.read_excel('AF-3024 Unit Cost Multivariate Regression.xlsx','UnitCostBreakdownData')
```

```python
ucbdf.head()
```

```python
ucbdf1 = ucbdf[['Time Period','AMN','QTY Output','Input Class','Input Type','Input','Cost Per Unit',
               'Unit Price','QTY Used']]
ucbdf1.head()
```

```python
ucbPM7398 = ucbdf1['AMN']==4137007398
ucbPM7398 = ucbdf1[ucbPM7398]
ucbPM7399 = ucbdf1['AMN']==4137007399
ucbPM7399 = ucbdf1[ucbPM7399]
ucbEC3025 = ucbdf1['AMN']==4191000580
ucbEC3025 = ucbdf1[ucbEC3025]
```

```python
#Function to calculate unit costs for each SKU based on time period
def unit_cost_sum(df):
    timeperiods = list(df['Time Period'].unique())
    unit_costs = []
    for i in timeperiods:
        df_it1 = df['Time Period']==i
        df_it2 = df[df_it1]
        unit_cost_it = round(df_it2['Cost Per Unit'].sum(),2)
        unit_costs.append(unit_cost_it)
    return unit_costs

uc_PM7398 = unit_cost_sum(ucbPM7398)
uc_PM7399 = unit_cost_sum(ucbPM7399)
uc_EC3025 = unit_cost_sum(ucbEC3025)
```

```python
#Create dictionary for each time period to unit cost value

def create_dic(uc,df):
    test_keys = list(df['Time Period'].unique())
    test_values = uc
    res = {test_keys[i]: test_values[i] for i in range(len(test_keys))} 
    return res

dic_PM7398 = create_dic(uc_PM7398,ucbPM7398)
dic_PM7399 = create_dic(uc_PM7399,ucbPM7399)
dic_EC3025 = create_dic(uc_EC3025,ucbEC3025)
```

```python
#Create new dataframe column Unit Cost with value for respective timeframe


# Define a function to map the values 
def set_value(row_number, assigned_value): 
    return assigned_value[row_number] 
event_dictionary = dic_PM7398
ucbPM7398['Unit Cost'] = ucbPM7398['Time Period'].apply(set_value, args =(event_dictionary, )) 

def set_value(row_number, assigned_value): 
    return assigned_value[row_number] 
event_dictionary = dic_PM7399
ucbPM7399['Unit Cost'] = ucbPM7399['Time Period'].apply(set_value, args =(event_dictionary, )) 

def set_value(row_number, assigned_value): 
    return assigned_value[row_number] 
event_dictionary = dic_EC3025
ucbEC3025['Unit Cost'] = ucbEC3025['Time Period'].apply(set_value, args =(event_dictionary, )) 
  
```

## PM-7398 Month-to-Month Cost Breakdown

```python
def unit_cost_pies(df):
    timeperiods = list(df['Time Period'].unique())
    for i in timeperiods:
        df_it1 = df['Time Period']==i
        df_it2 = df[df_it1]
        labels = list(df_it2['Input'])
        sizes = list(df_it2['Cost Per Unit'])
        if sizes[1] > 0:
            plt.figure(figsize = (25,10))
            plt.pie(sizes,labels=labels,shadow=True,startangle=90,autopct='%1.1f%%')
            plt.title('Unit Cost Breakdown for {}'.format(i))
            plt.show()
        else:
            continue
        
unit_cost_pies(ucbPM7398)
```

## PM-7399 Month-to-Month Cost Breakdown

```python
unit_cost_pies(ucbPM7399)
```

## EC-3025 Month-to-Month Cost Breakdown

```python
unit_cost_pies(ucbEC3025)
```
