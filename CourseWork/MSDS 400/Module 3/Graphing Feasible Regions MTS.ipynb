```python
import matplotlib.pyplot
from matplotlib.pyplot import *
import numpy
from numpy import *
```

```python
x = arange(-2,6.1,0.1) #This is another way to create a series of x values
#(start,end (no inclusive), iterate by)
y = -1.0+(1/4)*x
figure()
xlim(-2,6) #Set the x-limits for the graph/on the axes
ylim(-2,6)
xlabel('X-axis')
ylabel('Y-axis')
matplotlib.pyplot.hlines(0,-2,6,color = 'r')
matplotlib.pyplot.vlines(0,-2,6,color = 'r') #Plotting the axis onto the graph in a convenient way!
grid(True)
plot(x,y)
z=-2.0+0.0*x
plot(x,z)
fill_between(x,y,z,color='b')
title ('Shaded Area Shows the Linear Inequality x-4y > 4')
show()
```
